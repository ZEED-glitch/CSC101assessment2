```python
def load_transactions(filename):
    """Read lines from a transaction file safely."""
    try:
        with open(filename, 'r') as file:
            lines = file.readlines()
        return [line.strip() for line in lines if line.strip()]
    except FileNotFoundError:
        print(f"File '{filename}' not found!")
        return []

# Load transactions from the available file
transaction_data = load_transactions('/content/transactions.txt')

# Display the loaded transactions
print("Loaded Transactions:")
for transaction in transaction_data:
    print(transaction)
```

```python
def parse_transaction_line(line):
    """Parses a transaction line into a dictionary, handling corrupted lines.

    Args:
        line (str): A string representing a single transaction (e.g., 'Coffee,-4.50').

    Returns:
        dict or None: A dictionary {'item': str, 'amount': float} if successful,
                      otherwise None for corrupted lines.
    """
    try:
        parts = line.split(',')
        if len(parts) == 2:
            item = parts[0].strip()
            amount = float(parts[1].strip())
            return {'item': item, 'amount': amount}
        else:
            print(f"Warning: Skipping malformed line (incorrect number of parts): '{line}'")
            return None
    except ValueError:
        print(f"Warning: Skipping malformed line (could not convert amount to float): '{line}'")
        return None
    except Exception as e:
        print(f"An unexpected error occurred while parsing line '{line}': {e}")
        return None

# Apply the parsing function to the transaction_data
parsed_transactions = []
failed_lines_count = 0 # Initialize a counter for failed lines
for line in transaction_data:
    parsed = parse_transaction_line(line)
    if parsed:
        parsed_transactions.append(parsed)
    else:
        failed_lines_count += 1 # Increment counter if parsing failed

# Display the parsed transactions
print("\nParsed Transactions:")
for transaction in parsed_transactions:
    print(transaction)

# Report the number of failed lines
print(f"\nTotal failed lines: {failed_lines_count}")
```

```python

total_transactions = len(parsed_transactions)

# Initialize income and expenses
total_income = 0.0
total_expenses = 0.0

# Iterate through parsed transactions to calculate income and expenses
for transaction in parsed_transactions:
    amount = transaction['amount']
    if amount > 0:
        total_income += amount
    else:
        total_expenses += amount

# Display the results
print(f"\nTotal Transactions: {total_transactions}")
print(f"Total Income: {total_income:.2f}")
print(f"Total Expenses: {total_expenses:.2f}")
```
