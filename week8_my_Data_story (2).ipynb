```python
# This line defines the path to our file.
# The variable `file_path` now holds the location of 'my_data.txt'.
file_path = '/content/my_data.txt'

try:
    # The 'try' block attempts to execute the code within it.
    # If an error occurs, Python will look for an 'except' block to handle it.

    # This line opens the file specified by `file_path`.
    # 'r' means we open it in 'read' mode.
    # 'with open(...) as file:' is a safe way to open files;
    # it ensures the file is automatically closed after we're done, even if errors happen.
    # The opened file object is assigned to the variable `file`.
    with open(file_path, 'r') as file:
        # This line reads the entire content of the opened file.
        # The content (as a single string) is then stored in the `content` variable.
        content = file.read()
        # This line prints the `content` (which is the text from the file)
        # to the console, so you can see what was inside the file.
        print(content)

        # --- New functionality: Count lines and words ---

        # To count lines, we first split the content by newline characters.
        # .strip() removes any leading/trailing whitespace, including extra newlines,
        # and then we count the number of elements in the resulting list.
        num_lines = len(content.strip().split('\n'))

        # To count words, we first split the content by any whitespace.
        # This creates a list of words. We then count the number of elements in this list.
        # .split() without arguments handles multiple spaces and newlines as single delimiters.
        num_words = len(content.split())

        # Print the total number of lines found in the file.
        print(f"\nNumber of lines: {num_lines}")
        # Print the total number of words found in the file.
        print(f"Number of words: {num_words}")

# This 'except' block specifically catches a 'FileNotFoundError'.
# This happens if the file specified in `file_path` does not exist.
except FileNotFoundError:
    # If the file is not found, this message is printed to the console.
    # The `f-string` allows us to easily include the `file_path` in the message.
    print(f"Error: The file '{file_path}' was not found.")

# This is a general 'except' block that catches any other type of error
# that might occur during the 'try' block, other than `FileNotFoundError`.
# The error object itself is stored in the variable `e`.
except Exception as e:
    # If any other error occurs, this message is printed, along with the specific error details.
    print(f"An error occurred: {e}")
```
